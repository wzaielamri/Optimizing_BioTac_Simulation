## Calculate T-Test

```python
# t-tests:
from scipy import stats
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')


# code adjusted from https://gist.github.com/jensdebruijn/13e8eeda85eb8644ac2a4ac4c3b8e732
# calcualtion based on Bouckaert et al. 2004
from scipy.stats import t
from math import sqrt
from statistics import stdev

def corrected_dependent_ttest_TwoTailed(data1, data2, n_training_samples=226194, n_test_samples=29400, alpha=0.025):
    n = len(data1)
    differences = [(data1[i]-data2[i]) for i in range(n)]
    sd = stdev(differences)
    divisor = 1 / n * sum(differences)
    test_training_ratio = n_test_samples / n_training_samples  
    denominator = sqrt(1 / n + test_training_ratio) * sd 
    t_stat = divisor / denominator
    # degrees of freedom
    df = n - 1
    # calculate the critical value
    cv = t.ppf(1.0 - alpha, df)
    # calculate the p-value
    p = (1.0 - t.cdf(abs(t_stat), df)) * 2.0
    # return everything
    return t_stat, p #df, cv, 

def corrected_dependent_ttest_RightTailed(data1, data2, n_training_samples=226194, n_test_samples=29400, alpha=0.05):
    n = len(data1)
    differences = [(data1[i]-data2[i]) for i in range(n)]
    sd = stdev(differences)
    divisor = 1 / n * sum(differences)
    test_training_ratio = n_test_samples / n_training_samples  
    denominator = sqrt(1 / n + test_training_ratio) * sd 
    t_stat = divisor / denominator
    # degrees of freedom
    df = n - 1
    # calculate the critical value
    cv = t.ppf(1.0 - alpha, df)
    # calculate the p-value
    p = (1.0 - t.cdf(t_stat, df))
    # return everything
    return t_stat, p #df, cv, 


def corrected_dependent_ttest_LeftTailed(data1, data2, n_training_samples=226194, n_test_samples=29400, alpha=0.05):
    n = len(data1)
    differences = [(data1[i]-data2[i]) for i in range(n)]
    sd = stdev(differences)
    divisor = 1 / n * sum(differences)
    test_training_ratio = n_test_samples / n_training_samples  
    denominator = sqrt(1 / n + test_training_ratio) * sd 
    t_stat = divisor / denominator
    # degrees of freedom
    df = n - 1
    # calculate the critical value
    cv = t.ppf(1.0 - alpha, df)
    # calculate the p-value
    p = t.cdf(t_stat, df)
    # return everything
    return t_stat, p #df, cv, 
```
